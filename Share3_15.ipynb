```python
#FinalFunction
p3 = open("task3_final1.txt","r")
cont1 = p3.readlines()
for link in tqdm(cont1[280000:300000]):
    browser.get(link)
    int_html_page = browser.page_source
    int_page_soup = soup(int_html_page,"html.parser")
    
    name = int_page_soup.h1.text.split("\n")[0].strip()
        
    try:
        country = int_page_soup.h1.text.split("\n")[1].split("[")[1].split("]")[0]
    except:
        country = ""
        
    try:
        company_type = int_page_soup.findAll("div",{"id":"company_heading"})[0].text.split("\n")[2]
    except:
        company_type = ""
        
    #Projects In Development
    try:
        btn1 = browser.find_elements_by_xpath('//*[@id="a-autoid-0"]/span/input')[0]
        btn1.click()
        time.sleep(2)
        val = 57
        while(True):
            try: #//*[@id="in_development_filmography"]/tbody/tr[107]/td/div/span[1]/span/span/input
                time.sleep(2)
                btn2 = browser.find_element_by_xpath(f'//*[@id="in_development_filmography"]/tbody/tr[{val}]/td/div/span[1]/span/span/input')
                btn2.click()
                time.sleep(2)
                val = val+50
            except:
                break
    except:
        val = 57
        
    pid_html_page = browser.page_source
    pid_page_soup = soup(pid_html_page,"html.parser")
    try:
        res = pid_page_soup.findAll("table",{"id":"in_development_filmography"})[0].findAll("tr",{"class":"filmography_row"})
        pid_length = len(res)
    except:
        pid_length = 0
        pid_filmography1 = pid_filmography1.append([[name,company_type,country,"None","None","None","None"]])
    for i in range(pid_length):
        p_name = res[i].findAll("div",{"class":"a-section a-spacing-mini a-spacing-top-mini"})[0].text.split("\n")[3]
        p_meter = res[i].findAll("div",{"class":"a-section a-spacing-mini a-spacing-top-mini"})[1].text.strip("\n").strip()
        p_budget = res[i].findAll("div",{"class":"a-section a-spacing-mini a-spacing-top-mini"})[2].text.strip("\n")
        p_status = res[i].findAll("div",{"class":"a-section a-spacing-mini a-spacing-top-mini"})[3].text.strip("\n").strip(" ")
        pid_filmography1 = pid_filmography1.append([[name,company_type,country,p_name,p_meter,p_budget,p_status]])
        
    #Films In Development
    try:
        btn3 = browser.find_elements_by_xpath('//*[@id="a-autoid-1"]/span/input')[0]
        btn3.click()     
        time.sleep(2)
        val = 57
        while(True):
            try:
                time.sleep(2)
                btn4 = browser.find_element_by_xpath(f'//*[@id="in_production_film_filmography"]/tbody/tr[{val}]/td/div/span[1]/span/span/input')
                btn4.click()
                time.sleep(2)
                val = val+50
            except:
                break
    except:
        val = 57
        
    fid_html_page = browser.page_source
    fid_page_soup = soup(fid_html_page,"html.parser")
    try:
        res = fid_page_soup.findAll("table",{"id":"in_production_film_filmography"})[0].findAll("tr",{"class":"filmography_row"})
        fid_length = len(res)
    except:
        fid_length = 0
        fid_filmography1 = fid_filmography1.append([[name,company_type,country,"None","None","None","None"]])
    for i in range(fid_length):
        f_name = res[i].findAll("td")[0].text.split("\n")[3]
        f_meter = res[i].findAll("td")[1].text.strip("\n").strip()
        f_budget = res[i].findAll("td")[2].text.strip("\n").strip()
        f_status = res[i].findAll("td")[3].text.strip("\n").strip()
        fid_filmography1 = fid_filmography1.append([[name,company_type,country,f_name,f_meter,f_budget,f_status]])
        
    #Past Films and Videos
    try:
        btn5 = browser.find_element_by_xpath('//*[@id="a-autoid-2"]/span/input')
        btn5.click()
        time.sleep(2)
        val = 57
        while(True):
            try:
                time.sleep(2)
                btn6 = browser.find_element_by_xpath(f'//*[@id="past_film_filmography"]/tbody/tr[{val}]/td/div/span[1]/span/span/input')
                btn6.click()
                time.sleep(2)
                val = val + 50
            except:
                break
    except:
        val = 57
    
    pfv_html_page = browser.page_source
    pfv_page_soup = soup(pfv_html_page,"html.parser")
    try:
        res = pfv_page_soup.findAll("table",{"id":"past_film_filmography"})[0].findAll("tr",{"class":"filmography_row"})
        pfv_length = len(res)
    except:
        pfv_length = 0
        pfv_filmography1 = pfv_filmography1.append([[name,company_type,country,"None","None","None","None","None"]])
    for i in range(pfv_length):
        pn = res[i].findAll("td")[0].text.split("\n")[3]
        pb = res[i].findAll("td")[1].text.strip("\n").strip()
        powend = res[i].findAll("td")[2].text.strip("\n").strip()
        pg1 = res[i].findAll("td")[3].text.strip("\n").strip()
        pg2 = res[i].findAll("td")[4].text.strip("\n").strip()
        pfv_filmography1 = pfv_filmography1.append([[name,company_type,country,pn,pb,powend,pg1,pg2]])
    
    #Past Television    
    try:
        btn7 = browser.find_element_by_xpath('//*[@id="a-autoid-3"]/span/input')
        btn7.click()
        time.sleep(2)
        val = 57
        while(True):
            try:
                time.sleep(2)
                btn8 = browser.find_element_by_xpath(f'//*[@id="past_television_filmography"]/tbody/tr[{val}]/td/div/span[1]/span/span/input')
                btn8.click()
                time.sleep(2)
                val = val + 50
            except:
                break
    except:
        val = 57
        
    pt_html_page = browser.page_source
    pt_page_soup = soup(pt_html_page,"html.parser")
    try:
        res = pt_page_soup.findAll("table",{"id":"past_television_filmography"})[0].findAll("tr",{"class":"filmography_row"})
        tv_length = len(res)
    except:
        tv_length = 0
        tv_filmography1 = tv_filmography1.append([[name,company_type,country,"None","None","None"]])
    for i in range(tv_length):
        tvname = res[i].findAll("td")[0].text.split("\n")[3]
        meter = res[i].findAll("td")[1].text.strip("\n").strip()
        epi = res[i].findAll("td")[2].text.strip("\n").strip()
        tv_filmography1 = tv_filmography1.append([[name,company_type,country,tvname,meter,epi]])
        
pid_filmography1.to_csv("Pro_PID_Filmography15.csv",index=None)
fid_filmography1.to_csv("Pro_FID_Filmography15.csv",index=None)
pfv_filmography1.to_csv("Pro_PFV_Filmogrpahy15.csv",index=None)
tv_filmography1.to_csv("Pro_TV_Filmography15.csv",index=None)
```
