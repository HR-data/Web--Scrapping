```python
#print(len(t))
# Scraping

b=bs.findAll('ul',{"class":"title"})
Movie=[]
Year=[]
Director=[]
Production_comp=[]
Region_code=[]
Genre=[]
for i in b:
    Movie.append(i.find('span','display-title').find('a').text)
    Year.append(i.find('span','year').text.replace("(","").replace(")",""))
    Director.append(i.find("span","display-name").find('a').text)
    Production_comp.append(i.find("span","display-company").find('a').text)
    Region_code.append(i.find("span","region_code").text.replace('\n',"").strip().replace('[',"").replace(']',""))
    Genre.append(i.findAll('li','ellipsis')[4].text.strip()[7:])
    

Final_lst=[]

for m,y,d,p,r,g in zip(Movie,Year,Director,Production_comp,Region_code,Genre):
    Final_lst.append({'Movie':m,'Year':y,'Director':d,'Production_comp':p,'Region_code':r,'Genre':g})


df=pd.DataFrame(Final_lst)
df
```
