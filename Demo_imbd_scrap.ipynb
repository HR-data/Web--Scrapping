```python
from bs4 import BeautifulSoup 
import requests 
import re
import json
from selenium import webdriver
import pandas as pd
from urllib.request import urlopen as uReq

url='https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012'
resp = uReq(url) # where url is the above url 
page_html=resp.read()
resp.close()
bs = BeautifulSoup(page_html,'html.parser')
#print(r.text)
#bs = BeautifulSoup(r.text)
```

```python
t=bs.findAll("div",{"class":"lister-item-content"})
title=[]
certificate=[]
runtime=[]
genre=[]
ratings=[]
metascore=[]
details=[]
votes=[]
Gross=[]
Director=[]
t
```

```python
for movie in t:
    title.append(movie.find('a').contents[0])
    #certificate.append(movie.find("span","certificate").contents[0])
    runtime.append(movie.find("span","runtime").contents[0])
    genre.append(movie.find("span","genre").contents[0])
    ratings.append(movie.find("span","value").contents[0])
    metascore.append(movie.find("span","metascore").contents[0])
    for i in movie.findAll("p","text-muted"):
        details.append(i.contents[0])
    for i in movie.findAll("p","sort-num_votes-visible"):
        votes.append(i.contents[3].contents[0])
        Gross.append(i.contents[9].contents[0]) 
    lst=[i.contents[0] for i in movie.find("p","").findAll("a")] 
    Director.append(lst[0])
    stars=lst[1:]
    
for i in t:
    r=i.find("p","text-muted").contents[1]
    certificate.append(r.contents[0])   
    
final_lst=[]

for tit,cert,run,gen,rati,meta,det,vot,gro,di,sta in zip(title,certificate,runtime,genre,ratings,metascore,details,votes,Gross,Director,stars):
    final_lst.append({'Title':tit,'Certificate':cert,'Runtime':run,'Genre':gen,'Ratings':rati,'Metascore':meta,
                      'Details':det,'Votes':vot,'Gross':gro,'Director':di,'Stars':sta})

df=pd.DataFrame(final_lst)
df
    #print(title,certificate,runtime,genre,ratings,metascore,detail,Director,stars,votes,Gross)    
```
